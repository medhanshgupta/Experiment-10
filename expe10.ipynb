```python
# create DataFrame
import pandas as pd
data = {"Name":["A","B","C"],"Marks":[85,90,78]}
df = pd.DataFrame(data)
print(df)
```

```python
# access particular element
print(df.loc[0,"Name"])
print(df.iloc[2,1])
```

```python
# Add new column

df["Grade"] = ["First Class","Distinction","Second Class"]
print(df)
```

```python
# update data in dataframe
# change marks of student A to 88
df.loc[0,"Marks"] = 88
print(df)
```

```python
# delete a column
# df.drop("Grade",axis=1,inplace=True)
df1 = df.drop("Grade",axis=1)
print(df)
print(df1)
```

```python
# perform basic statistical analysis
mean1=df["Marks"].mean()
max1 = df["Marks"].max()
min1=df["Marks"].min()
print(mean1)
print(min1)
print(max1)
```

```python
# apply condition (filtering)
# display students scoring more than 80.
df[df["Marks"]>80]
```
